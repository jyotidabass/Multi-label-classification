```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
att = list(df.columns)
```

```python
text_att = att[:-8]
```

```python
text_att[:10]
```

```python
unigram, bigram, trigram = {}, {}, {}
```

```python
for a in text_att:
    length = len(a.split('_'))
    if length == 1:
        unigram[a] = 0
    elif length == 2:
        bigram[a] = 0
    elif length == 3:
        trigram[a] = 0
```

```python
len(unigram) + len(bigram) + len(trigram)
```

```python
len(text_att)
```

```python
for k in unigram.keys():
    unigram[k] = sum(df[k])
for k in bigram.keys():
    bigram[k] = sum(df[k])
for k in trigram.keys():
    trigram[k] = sum(df[k])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(background_color='white', margin=0)
```

```python
unigram_sorted = sorted(unigram.items(), key = lambda t: t[1], reverse=True)
```

```python
plot_unigram = unigram_sorted[:40]
```

```python
pic1 = wc.generate_from_frequencies(plot_unigram)
```

```python
plt.imshow(pic1)
plt.axis('off')
plt.savefig('unigram_wordcloud.png')
plt.show()
```

```python
bigram_sorted = sorted(bigram.items(), key = lambda t: t[1], reverse=True)
```

```python
plot_bigram = bigram_sorted[:40]
pic2 = wc.generate_from_frequencies(plot_bigram)
plt.imshow(pic2)
plt.axis('off')
plt.savefig('bigram_wordcloud.png')
plt.show()
```

```python
trigram_sorted = sorted(trigram.items(), key = lambda t: t[1], reverse=True)
```

```python
plot_trigram = trigram_sorted[:30]
pic3 = wc.generate_from_frequencies(plot_trigram)
plt.imshow(pic3)
plt.axis('off')
plt.savefig('trigram_wordcloud.png')
plt.show()
```
