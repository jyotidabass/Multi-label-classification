```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df_y = df.loc[:, ['IsFoodGood', 'IsServiceGood', 'IsAmbianceGood', 'IsDealsGood','IsPriceGood']]
```

```python
labels = list(df_y.columns)
```

```python
label_cnt = {}
for label in labels:
    label_cnt[label] = sum(df_y[label])
```

```python
label_cnt
```

```python
result = sorted(label_cnt.items(), key = lambda t: t[1], reverse=True)
```

```python
label, cnt = zip(*result)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
colors = ('#FCFA6A', '#E6725A', '#A0FDF2', '#C49CFD', '#569AFC')
plt.pie(cnt, labels=label, labeldistance=1.1,
        startangle=90, autopct='%3.1f%%', pctdistance=0.6, colors=colors)
plt.axis('equal')
plt.savefig('label_dist.png')
plt.show()
```
