```python
import arff
import pandas as pd
```

```python
arff_d = arff.load('train.arff')
```

```python
attributes = []
with open('train.arff') as f:
    for line in f:
        if not line.startswith('@'):
            break
        if line.startswith('@attribute'):
            a = line.split(' ')[1]
            attributes.append(a)
```

```python
data = pd.DataFrame(arff_d, columns=attributes)
```

```python
data
```

```python
data.to_csv('train.csv', header=attributes, index=False)
```
